```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr

# 读取数据
df = pd.read_excel('dataset_cn.xlsx', sheet_name=0)

# 获取第4列至第23列（即第4至第23列的所有回答数据）
# 记得：Excel中的列索引从0开始，所以第4列到第23列对应的索引是3到22
responses = df.iloc[:, 3:23]

# 定义计算Cronbach's Alpha的函数
def cronbach_alpha(data):
    # 计算每个题目的方差
    item_variances = data.var(axis=0, ddof=1)
    print(item_variances)
    
    # 计算总分（每行的答案之和）
    total_score = data.sum(axis=1)
    
    # 计算总分的方差
    total_variance = total_score.var(ddof=1)
    
    # Cronbach's Alpha公式
    alpha = (data.shape[1] / (data.shape[1] - 1)) * (1 - item_variances.sum() / total_variance)
    
    return alpha

# 计算Cronbach's Alpha
alpha = cronbach_alpha(responses)

print(f"Cronbach's Alpha值: {alpha}")
```

```python
import numpy as np
from scipy.stats import pearsonr

# 示例数据（Q2 和 Q5 的 5 个数值）
Q2 = [20, 42, 50, 22, 9]
Q5 = [12, 43, 45, 28, 15]

# 计算皮尔逊相关系数和 p 值
r, p = pearsonr(Q2, Q5)

# 输出结果
print(f"皮尔逊相关系数 (r): {r}")
print(f"p 值: {p}")

# 检验统计显著性
if p < 0.01:
    print("结果显著（p < 0.01），可以认为两个变量之间存在显著的正相关关系。")
else:
    print("结果不显著，不能认为两个变量之间有显著的正相关关系。")
```
